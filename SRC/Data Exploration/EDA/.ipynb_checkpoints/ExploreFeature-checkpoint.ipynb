```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_wine

%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = (22,22)
```

# Read the data 

```python
df_train_est = pd.read_csv('train_est_A.csv')
df_train_est_A = df_train_est.rename(columns={'Unnamed: 0': 'date_forecast'})
df_train_obs = pd.read_csv('train_obs_A.csv')
df_train_obs_A = df_train_obs.rename(columns={'Unnamed: 0': 'date_forecast'})
df_test = pd.read_csv('X_test_A.csv')
df_test_A = df_test.rename(columns={'Unnamed: 0': 'date_forecast'})
df_train_est_A.set_index('date_forecast', inplace=True)

df_train_obs_A.set_index('date_forecast', inplace=True)

df_test_A.set_index('date_forecast', inplace=True)
```

# Train/test split 

```python
from sklearn.model_selection import train_test_split

X = df_train_est_A.drop(['pv_measurement'], axis=1)
y = df_train_est_A['pv_measurement']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle = False)
```

```python
print(X_train)
print(y_train)
```

# Baseline model: Gradient Boosting Classifier with all features 

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error

# Initialize classifier
gbc = GradientBoostingRegressor(max_depth=5, random_state=42)

# Train classifier using all features
gbc.fit(X_train, y_train)

# Make predictions
preds = gbc.predict(X_test)

# Evaluate the model using the F1-score
mae = mean_absolute_error(y_test, preds)

print(mae)
```

# Feature selection techniques

```python
X_train_v1, X_test_v1, y_tain_v1, y_test_v1 = X_train.copy(), X_test.copy(), y_train.copy(), y_test.copy()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled_X_train_v1 = scaler.fit_transform(X_train_v1)
```

```python
sel_X_train_v1 = X_train_v1.drop(['elevation:m', 'snow_drift:idx'], axis=1)
sel_X_test_v1 = X_test_v1.drop(['elevation:m', 'snow_drift:idx'], axis=1)

gbc.fit(sel_X_train_v1, y_train)

var_preds = gbc.predict(sel_X_test_v1)

# Evaluate the model using the F1-score
mae_var = mean_absolute_error(y_test, preds)

print(mae_var)
```

```python
fig, ax = plt.subplots()

x = ['All features', 'Variance threshold']
y = [mae, mae_var]

ax.bar(x, y, width=0.2)
ax.set_xlabel('Feature selection method')
ax.set_ylabel('F1-Score (weighted)')

for index, value in enumerate(y):
    plt.text(x=index, y=value+0.01, s=str(round(value,3)), ha='center')
    
plt.tight_layout()
```

## K-best features 

```python
X_train_v2, X_test_v2, y_train_v2, y_test_v2 = X_train.copy(), X_test.copy(), y_train.copy(), y_test.copy()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

mae_kbest_list = []

for k in range(1, X_train_v2.shape[1]):
    selector = SelectKBest(f_regression, k=k)
    selector.fit(X_train_v2, y_train_v2)
    
    sel_X_train_v2 = selector.transform(X_train_v2)
    sel_X_test_v2 = selector.transform(X_test_v2)
    
    gbc.fit(sel_X_train_v2, y_train_v2)
    kbest_preds = gbc.predict(sel_X_test_v2)
    
    mae_kbest = mean_absolute_error(y_test, preds)
    
    mae_kbest_list.append(mae_kbest)
```

```python
fig, ax = plt.subplots()

x = np.arange(1, X_train_v2.shape[1])
y = mae_kbest_list

ax.bar(x, y, width=0.2)
ax.set_xlabel('Number of features selected using mutual information')
ax.set_ylabel('Mae(weighted)')
ax.set_xticks(np.arange(1, 14))
ax.set_xticklabels(np.arange(1, 14), fontsize=12)

for i, v in enumerate(y):
    plt.text(x=i+1, y=v+0.05, s=str(v), ha='center')
    
plt.tight_layout()
```

```python
selector = SelectKBest(f_regression, k=3)
selector.fit(X_train_v2, y_train_v2)

selected_feature_mask = selector.get_support()

selected_features = X_train_v2.columns[selected_feature_mask]

selected_features
```

## Recursive feature elimination (RFE) 

```python
X_train_v3, X_test_v3, y_train_v3, y_test_v3 = X_train.copy(), X_test.copy(), y_train.copy(), y_test.copy()
```

```python
from sklearn.feature_selection import RFE

mae_rfe_list = []

for k in range(1, X_train_v2.shape[1]):
    RFE_selector = RFE(estimator=gbc, n_features_to_select=k, step=1)
    RFE_selector.fit(X_train_v3, y_train_v3)
    
    sel_X_train_v3 = RFE_selector.transform(X_train_v3)
    sel_X_test_v3 = RFE_selector.transform(X_test_v3)
    
    gbc.fit(sel_X_train_v3, y_train_v3)
    RFE_preds = gbc.predict(sel_X_test_v3)
    
    mae_rfe = mean_absolute_error(y_test, preds)
    print("iteration ",k," mae: ", mae_rfe)
    mae_rfe_list.append(mae_rfe)
```

```python
fig, ax = plt.subplots()

x = np.arange(1, X_train_v2.shape[1])
y = rfe_f1_score_list

ax.bar(x, y, width=0.2)
ax.set_xlabel('Number of features selected using RFE')
ax.set_ylabel('F1-Score (weighted)')
ax.set_xticks(np.arange(1, 14))
ax.set_xticklabels(np.arange(1, 14), fontsize=12)

for i, v in enumerate(y):
    plt.text(x=i+1, y=v+0.05, s=str(v), ha='center')
    
plt.tight_layout()
```

```python
RFE_selector = RFE(estimator=gbc, n_features_to_select=3, step=10)
RFE_selector.fit(X_train_v3, y_train_v3)

selected_features_mask = RFE_selector.get_support()

selected_features = X_train_v3.columns[selected_features_mask]
selected_features
```
