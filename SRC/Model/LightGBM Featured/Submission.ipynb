```python
import pandas as pd
A = pd.read_csv('A.csv')
B = pd.read_csv('B.csv')
C = pd.read_csv('C.csv')
```

```python
pred = pd.concat([A,B,C], ignore_index = True)
```

```python
pred = pred.drop(columns = 'Unnamed: 0')
```

```python
pred['id'] = range(2160)
```

```python
pred = pred.set_index(pred['id'])
pred = pred.drop(columns = 'id', axis = 1)
pred
```
